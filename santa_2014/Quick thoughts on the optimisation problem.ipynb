```python
toys = np.load("toys_rev2.npy")
print(len(toys))
toys
```

Because of the rest period rule for unsanctioned working, elves can effectively only work 10 hours a day, with a fluctuating productivity rating.

```python
print("Assuming all 900 elves work at productivity 4.0 10 hours a day (impossible...) this would take...")
# Summing overflows 32-bit integers!
times = toys[:,1].astype(np.int64)
total_mins = np.sum(times)
print("Total work: {} minutes".format(total_mins))
print("Max and min times:", np.max(toys[:,1]), "--",np.min(toys[:,1]))
print("Crude lower bound: {} minutes = {:.3f} years".format(total_mins//(900*4),
        total_mins/(900*4)/(365*10*60)))
```

Remember that the sample "naive" solution takes 275700552 minutes = 524.5 years.

This hints at what I think will be important: how do we keep the Elves productivity up?  You clearly would wish to have a maxed out elf attack a toy which takes 47473 minutes to finish.  But this will take the Elf 11869 minutes, or 8 days, 5 hours and 49 minutes.  Starting at 9am, the elf would use `14*8=112` unsanctioned hours, and `5149 / 60` sanctioned hours, so new productivity will be:
$$ 4.0 \times 1.02^{5149/60} \times 0.9^{112} = 1.64\times 10^{-4} $$
that is, capped at $0.25$.  Furthermore, the poor elf is now allocated 112 hours of rest, taking over 11 days!  Our lower bound is extremely crude!  After that, we presumably would want to try to get this elf's productivity back up.

However, starting at $0.25$ we need $n$ hours of work to get back to $4.0$ which is $4.0 = 0.25\times 1.02^n$ and so $\log(16) = n \log(1.02)$ so $n \approx 140$ or 2 weeks of work!
