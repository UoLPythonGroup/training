```python
import pandas as pd
import numpy as np
```

Let's try to use Pandas to load the data set!  (Copying Jitse...)

```python
toys = pd.read_csv("toys_rev2.csv", index_col="ToyId",
                    parse_dates=["Arrival_time"])
toys
```

```python
print(toys.Arrival_time[1], type(toys.Arrival_time[1]))
```

So that doesn't work: the dates aren't automatically parsed.

Stackoverflow comes to our rescue: http://stackoverflow.com/questions/17465045/can-pandas-automatically-recognize-dates

(I used `datetime` before...)

```python
dateparse = lambda x: pd.datetime.strptime(x, '%Y %m %d %H %M')

toys = pd.read_csv("toys_rev2.csv", index_col="ToyId",
                    parse_dates=["Arrival_time"], date_parser = dateparse)
toys
```

```python
print(toys.Arrival_time[1], type(toys.Arrival_time[1]))
```

That works, but it's slow...  (no real surprise).  Let's see what the type of the dates are...

```python
dt = toys.Arrival_time[1]
dt
```

This is a pandas datatype.  It support arithmetic is an intelligent way:

```python
dt - pd.datetime(2014,1,1,0,0,0)
```

The following is pleasing fast!

```python
toys.Arrival_time -= pd.datetime(2014,1,1,0,0,0)
```

```python
toys.Arrival_time
```

```python
toys.Arrival_time[1]
```

Ah, so we have now a numpy datatype!  This also supports arithmetic:

```python
toys.Arrival_time[1] / np.timedelta64(1,"m")
```

```python
toys.Arrival_time /= np.timedelta64(1,"m")
```

```python
toys.Arrival_time = toys.Arrival_time.astype(np.int32)
```

```python
toys.to_pickle("toys_rev2.pd")
```
