# For sheet 4:

Plot graphs of $f(x) = \cosh(x)$ and $g(x) = \tanh(x)$ etc

```python
f = cosh(x)
g = tanh(x)
fprime = diff(f,x)
gprime = diff(g,x)

x_0 = 1

f_slope = fprime.subs(x,x_0)
g_slope = gprime.subs(x,x_0)

f_tangent = f_slope*x 
g_tangent = g_slope*x

f_intercept = f.subs(x,x_0)-f_slope*x_0
g_intercept = f.subs(x,x_0)-f_slope*x_0

plot(f,f_tangent+f_intercept,(x,-1,3))
plot(g,g_tangent+g_intercept,(x,-1,3))
```

```python
t = Symbol("t")
y = Function("y")

eq1 = Eq(diff(y(t),t,t)+2*diff(y(t),t)+5*y(t),0)
sol1=dsolve(eq1,y(t)).rhs

const1eq = Eq(sol1.subs(t,0),3)
const2eq = Eq(diff(sol1,t).subs(t,0),-5)

C1, C2 = symbols('C1 C2')

const1, const2 = solve([const1eq,const2eq])[C1],solve([const1eq,const2eq])[C2]

sol1ivp = sol1.subs(C1,const1).subs(C2,const2)

plot(sol1ivp,(t,-1,6))
```

```python
eq2 = Eq(diff(y(t),t,t)+2*diff(y(t),t)+5*y(t),3*sin(2*t))
sol2=dsolve(eq2,y(t)).rhs

const1eq = Eq(sol2.subs(t,0),3)
const2eq = Eq(diff(sol2,t).subs(t,0),-5)

C1, C2 = symbols('C1 C2')

const1, const2 = solve([const1eq,const2eq])[C1],solve([const1eq,const2eq])[C2]

sol2ivp = sol2.subs(C1,const1).subs(C2,const2)

plot(sol2ivp,(t,0,15))
```

```python
eq3 = Eq(diff(y(t),t,t) + diff(y(t),t) + 2*y(t),t**2*cos(4*t))
sol3=dsolve(eq3,y(t)).rhs

const1eq = Eq(sol3.subs(t,0),0)
const2eq = Eq(diff(sol3,t).subs(t,0),0)
C1, C2 = symbols('C1 C2')
const1, const2 = solve([const1eq,const2eq])[C1],solve([const1eq,const2eq])[C2]
sol3ivp = sol3.subs(C1,const1).subs(C2,const2)

plot(sol3ivp,(t,0,12))
```
